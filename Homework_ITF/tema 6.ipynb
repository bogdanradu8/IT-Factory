```python
# 1. 
# Clasa Cerc

# Atribute: raza, culoare

# Constructor pt ambele atribute

# Metode:
# descrie_cerc() - va PRINTA culoarea si raza
# aria() - va RETURNA aria 
# diametru() 
# circumferinta()

from math import pi

class Cerc:
    
    def __init__(self, raza, culoare):
        self.raza = float(raza)
        self.culoare = culoare
    
    def descrie_cerc(self):
        print(f'Cercul are raza {self.raza} si culoarea {self.culoare}')

    def aria(self):
        return  pi * self.raza ** 2
    
    def diametru(self):
        return self.raza * 2
    
    def circumferinta(self):
        return 2 * pi * self.raza

my_cerc = Cerc(21, "rosu")

print(my_cerc.aria())
print(my_cerc.circumferinta())
print(my_cerc.diametru())
my_cerc.descrie_cerc()
```

```python
# 2. 
# Clasa Dreptunghi

# Atribute: lungime, latime, culoare

# Constructor pt toate atributele

# Metode:
# descrie()
# aria()
# perimetrul()
# schimba_culoarea(noua_culoare) - aceasta metoda nu returneaza nimic. 
# Ea va lua ca si param o noua culoare si va suprascrie atributul self.culoare. Puteti verifica schimbarea culorii prin apelarea metodei descrie().

class Dreptunghi:
    def __init__(self, lungime, latime, culoare):
        self.lungime = lungime
        self.latime = latime
        self.culoare = culoare
    
    def descrie(self):
        print(f'Dreptunghiul are culoarea {self.culoare}, lungimea {self.lungime} si latimea {self.latime}')

    def aria(self):
        return self.lungime * self.latime
    
    def perimetrul(self):
        return 2 * self.lungime + 2 * self.latime

    def schimba_culoarea(self, culoare_new):
        self.culoare = culoare_new

my_drept = Dreptunghi(10, 5, 'rosu')
my_drept.descrie()
print(my_drept.aria())
print(my_drept.perimetrul())
my_drept.schimba_culoarea('verde')
my_drept.descrie()
```

```python
# 3.
# Clasa Angajat

# Atribute: nume, prenume, salariu

# Constructor pt toate atributele

# Metode:
# descrie()
# nume_complet()
# salariu_lunar()
# salariu_anual()
# marire_salariu(procent)

class Angajat:
    def __init__(self, nume ,prenume, salariu = 0.0):
        self.nume = nume
        self.prenume = prenume
        self.salariu = salariu
    
    def descrie(self):
        print(f'Angajatul are numele {self.nume}, prenumele {self.prenume} si are salariul de {self.salariu} RON.')
    
    def nume_complet(self):
        print(f'Numele complet este: {self.nume} {self.prenume}')
    
    def salariu_lunar(self):
        print(f'Salariul lunar = {self.salariu}')
    
    def salariu_anual(self):
        print(f'Salariul anual este {self.salariu * 12}')

    def marire_salariu(self, procent = 0.0):
        print(f'Marirea de {procent}% este de {self.salariu * procent / 100} RON')

a1 = Angajat('Popescu', 'Ion', 4000)

a1.descrie()
a1.nume_complet()
a1.salariu_lunar()
a1.salariu_anual()
a1.marire_salariu(30.5)
```

```python
# 4.
# Clasa Cont

# Atribute: iban, titular_cont, sold

# Constructor pentru toate

# Metode:
# afisare_sold() - Titularul x are in contul y suma de n lei
# debitare_cont(suma)
# creditare_cont(suma)

class Cont:
    def __init__(self, iban, titular_cont, sold = 0.0):
        self.iban = iban
        self.titular_cont = titular_cont
        self.sold = sold

    def afisare_sold(self):
        print(f'Soldul este {self.sold}')

    def debitare_cont(self, amount):
        if amount > self.sold:
            print('Fonduri insuficiente!')
                           
        else:
            self.sold -= amount    
            print(f'Contul a fost debitat cu {amount}')             
    
    def creditare_cont(self, amount):
        self.sold += amount
        return amount

client = Cont('RO55BTRL20120030X', 'Andrei')

client.afisare_sold()
client.debitare_cont(100)
client.creditare_cont(200)
client.afisare_sold()
client.debitare_cont(100)
```

```python
# 5.
# Clasa Factura

# Atribute: seria (va fi constanta, nu trebuie constructor, toate facturile vor avea aceeasi serie), numar, nume_produs, cantitate, pret_buc

# Constructor: toate atributele, fara serie

# Metode:
# schimba_cantitatea(cantitate)
# schimba_pretul(pret)
# schimba_nume_produs(nume)
# genereaza_factura() - va printa tabelar daca reusiti

# Factura seria x numar y
# Data: generati automat data de azi
# Produs | cantitate | pret bucata | Total 
# Telefon |      7       |       700       | 49000
from datetime import date

 
class Factura:    
    serie_fact = "223"
    def __init__(self, numar, nume_produs, cantitate, pret_buc):        
        self.numar = numar
        self.nume_produs = nume_produs
        self.cantitate = cantitate
        self.pret_buc = pret_buc
        
    def schimba_cantitatea(self, cantitate_noua):
        self.cantitate = cantitate_noua
    
    def schimba_pretul(self, pret_nou):
        self.pret_buc = pret_nou
    
    def schimba_nume_produs(self, nume_nou):
        self.nume_produs = nume_nou
    
    def genereaza_factura(self):
        
        print(f'Factura seria {Factura.serie_fact} nr {self.numar}')
        print(f'Data: {date.today()}')
        print('Produs | Cantitate | Pret/Buc | Total')
        print(f'{self.nume_produs} | {self.cantitate} | {self.pret_buc} | {self.pret_buc * self.cantitate}')
    
a = Factura(23, 'telefon', 7, 700)

a.genereaza_factura()
```

```python
# 6.
# Clasa Masina

# Atribute: marca, model, viteza maxima, viteza_actuala, culoare, culori_disponibile (set), inmatriculata (bool)
# Culoare = gri - toate masinile cand ies din fabrica sunt gri
# Viteza_actuala = 0 - toate masinile stau pe loc cand ies din fabrica
# Culori disponibile = alegeti voi 5-7 culori
# Marca = alegeti voi - fabrica produce o singura marca dar mai multe modele
# Inmatriculata = False

# Constructor: model, viteza_maxima

# Metode:
# descrie() (se vor printa toate atributele, inafara de culori_disponibile)
# inmatriculare() - va schimba atributul inmatriculata in True
# vopseste(culoare) - se va vopsi masina in noua culoare DOAR daca noua culoare e in optiunea de culori displonibile, altfel afisati o eroare
# accelereaza(viteza) - se va accelera la o anumiota viteza, daca viteza e negativa-eroare, daca viteza e mai mare decat viteza_max - masina va accelera pana la viteza maxima
# franeaza() - masina se va opri si va avea viteza 0

class Masina:
    def __init__(self, model, viteza_maxima):
        self.marca = 'Toyota'
        self.model = model
        self.viteza_max  = viteza_maxima
        self.viteza_actuala = 0 
        self.culoare = 'alba'
        self.culori_disp = {'alba', 'neagra', 'rosie', 'galbena', 'verde'}
        self.inmatriculata = False

    def descrie(self):
        print(f'''Masina marca {self.marca}, modelul {self.model}, prinde viteza maxima de {self.viteza_max} km/h si are culoarea {self.culoare}\n'''
              f'''Are viteza actuala {self.viteza_actuala} km/h. Este inmatriculata? {'DA' if self.inmatriculata else 'Nu'}''')
    
    def inmatriculare(self):
        self.inmatriculata = True
        print('Masina este acum inmatriculata!')
    
    def vopseste(self, culoare_nou):
        if culoare_nou in self.culori_disp:
            self.culoare = culoare_nou
            print(f'Ati vopsit masina. Acum are culoarea: {self.culoare}')
        else:
            print('Nu avem aceasta culoare')

    def accelereaza(self, viteza_nou):
        if viteza_nou < 0:
            print('Nu poate accelera')
        else:
            if viteza_nou > self.viteza_max:
                self.viteza_actuala = self.viteza_max
            else:
                self.viteza_actuala = viteza_nou
        print(f'Yuhuuu!, viteza actuala este {self.viteza_actuala} km/h')
    
    def franeaza(self):
        self.viteza_actuala = 0
        print('Oprim masina')

my_car = Masina('Auris', 200)
my_car.descrie()
my_car.vopseste('rosie')
my_car.accelereaza(400)
my_car.inmatriculare()
my_car.franeaza()
```

```python
# 7. Clasa TodoList
 
# Atribute: todo (dict, cheia e numele taskului, valoarea e descrierea)
# La inceput nu avem taskuri, dict e gol si nu avem nevoie de constructor

# Metode:
# adauga_task(nume, descriere) - se va adauga in dict
# finalizeaza_task(nume) - se va sterge din dict
# afiseaza_todo_list() - doar cheile
# afiseaza_detalii_suplimentare(nume_task) - in f de numele taskului, printam detalii suplimentare
# daca taskul nu e in todo list, intrebam utilizatorul daca vrea sa il adauge.
# Daca acesta raspunde nu - la revedere
# daca raspunde da - il cerem detalii task si salvam nume+detalii in dict

class TodoList:
    def __init__(self):
        self.todo = {}

    def adauga_task(self, nume, descriere):
        self.todo[nume] = descriere

    def finalizeaza_task(self, nume):
        self.todo.pop(nume)

    def afiseaza_todo_list(self):
        print(list(self.todo.keys()))

    def afiseaza_detalii(self, nume_task):
        if nume_task not in self.todo:
            adaugare = input(f'Vrei sa adaugi task-ul {nume_task}? Da sau Nu: ')
            if adaugare.lower() == 'da':
                detalii = input('Adauga detalii despre task: ')
                self.todo[nume_task] = detalii
            else:
                print('La revedere.')
        else:
            print(f'Task-ul {nume_task},\nDescriere: {self.todo[nume_task]}')


lista_de_rezolvat = TodoList()
lista_de_rezolvat.afiseaza_todo_list()
lista_de_rezolvat.adauga_task('repara masina', 'Trebuie sa schimbi electromotorul!')
lista_de_rezolvat.afiseaza_todo_list()
lista_de_rezolvat.afiseaza_detalii('repara masina')
lista_de_rezolvat.afiseaza_detalii('repara bicicleta')
lista_de_rezolvat.afiseaza_detalii('repara bicicleta')
```
